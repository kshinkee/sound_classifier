```python
from micromlgen import port
from sklearn.svm import SVC
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
clf = SVC(kernel='rbf', gamma=0.001).fit(X, y)
```

```python
print(port(clf, classmap={
    0: 'setosa',
    1: 'virginica',
    2: 'versicolor'
}))
```
